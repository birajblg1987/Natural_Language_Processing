```python
# import libraries
import string
import numpy as np 
from matplotlib import pyplot as plt 
from collections import Counter
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
# import data
text = open('Transcript.txt', encoding = 'utf-8', errors='ignore').read()

# convert all to lowercase
lower_case = text.lower()

# remove punctuations
cleaned_text = lower_case.translate(str.maketrans('','',string.punctuation))
```

```python
# tokenize words
tokenized_words = word_tokenize(cleaned_text,'english')
```

```python
# removing stop workds
final_words = []
for word in tokenized_words:
    if word not in stopwords.words('english'):
        final_words.append(word)
```

```python
# sentiment analysis
def sentiment_analyse(sentiment_text):
    score = SentimentIntensityAnalyzer().polarity_scores(sentiment_text)
    Negative = score['neg']
    Positive = score['pos']
    Neutral  = score['neu']
    return Negative, Positive, Neutral
```

```python
# call the function
Setiment_score = sentiment_analyse(cleaned_text)
```

```python
# creating dataset 
labs = ['Negative', 'Positive', 'Neutral']  
data = Setiment_score 

# Creating plot 
fig = plt.figure(figsize =(4,47)) 
plt.pie(data, labels = labs) 

plt.title('Facebook Q2 Earnings Call Sentiment')

# show plot 
plt.show() 
```

# Word Cloud

```python
# update stopwords with additional words from us
my_stopwords = set(STOPWORDS)
my_stopword_list = ['billion', 'we', 're', 'facebook', 'us']

for i in my_stopword_list:
    my_stopwords.add(i)
```

```python
textcloud = WordCloud(stopwords = my_stopwords, relative_scaling=1.0).generate(cleaned_text) 
```

```python
plt.figure(figsize = (10,7))
plt.imshow(textcloud)
plt.axis("off")
plt.show()
```
